```python
import networkx as nx

# Create a directed networkx graph
G = nx.DiGraph()

# Add web pages (graph nodes)
for i in range(4):
   G.add_node(i, label="p" + str(i))

# Add outgoing web links with weights (directed graph edges)
G.add_edge(0, 1, weight=1.0/3.0, label="1/3")
G.add_edge(0, 2, weight=1.0/3.0, label="1/3")
G.add_edge(0, 3, weight=1.0/3.0, label="1/3")

G.add_edge(1, 2, weight=1.0/2.0, label="1/2")
G.add_edge(1, 3, weight=1.0/2.0, label="1/2")

G.add_edge(2, 0, weight=1.0, label="1")

G.add_edge(3, 0, weight=1.0/2.0, label="1/2")
G.add_edge(3, 2, weight=1.0/2.0, label="1/2")

# To plot graph, convert to a PyGraphviz graph for drawing
Ag = nx.to_agraph(G)
Ag.layout(prog='dot')
Ag.draw('test.png')

# Display network of web pages
from IPython.display import Image
Image('test.png')
```

```python
# Get the matrix containing the graph weights
A = nx.adjacency_matrix(G)
print A.T
```
